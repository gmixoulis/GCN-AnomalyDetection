```python
# The file with features is anonimyzed, it doesn't even have column names.
#Let's see what can be derived from the data description:

#1) The first column (column 0) is the transaction id

#2) Column 1 represents timesteps for each node. These timestamps have an interval of ~2 weeks. 
#   Each timestamp contains connected component of transactions,
#   which appeared on the blockchain within less than three hours between each other

#3) Next 93 features represent local information about the transaction — including the time step described above, 
#   number of inputs/outputs, transaction fee, output volume and aggregated figures 
#   such as average BTC received (spent) by the inputs/outputs and average number 
#   of incoming (outgoing) transactions associated with the inputs/outputs.

#4) The  72 features represent nonlocal (graph) information in the form of aggregated features,
#   obtained using transaction information one-hop backward/forward from the center node — giving the maximum
#   ,minimum, standard deviation and correlation coefficients of the neighbour transactions 
#   for the same information data (number of inputs/outputs, transaction fee, etc.)

# so we will rename our features acoordingly 
df_features.columns = ['ID', 'Time step'] + [f'Trans_feature_{i}' for i in range(93)] + [f'Graph_feature_{i}' for i in range(72)] 
```

```python
# so we should explore a little bit better the form of this connected components in some time steps and to 
# visualize the graphs that are created

# lets start with timestamp 10 and illicit transactions (class 1 )
# we will create first the directed graph that is formed in this case
# first we find the illicit in timestep 10 
illicit = df_class_and_features.loc[(df_class_and_features['Time step'] == 10) & (df_class_and_features['Class'] == '1')]
print(illicit)
```

```python
# we will first visualize them with a directed graph 
G = nx.from_pandas_edgelist(edges, source = 'ID1', target = 'ID2', 
                                 create_using = nx.DiGraph())

#Position nodes using Fruchterman-Reingold force-directed algorithm.
#The algorithm simulates a force-directed representation of the network treating edges 
#as springs holding nodes close, while treating nodes as repelling objects, 
#sometimes called an anti-gravity force. Simulation continues until the positions are close to an equilibrium
pos = nx.spring_layout(G)
# print(pos)
#without labels 
# nx.draw(G, pos = pos)
#with labels
nx.draw(G, pos = pos, with_labels = True)
# nx.draw(graph, cmap = plt.get_cmap('terrain'), with_labels=True, pos=pos)

# As we can see from the graph not even one user had illicit transactions with more than one receiver 
```

```python
# now we will visualize the corresponding undirected graph
G = nx.from_pandas_edgelist(edges, source = 'ID1', target = 'ID2', 
                                 create_using = nx.Graph())
pos = nx.spring_layout(G)
# print(pos)
nx.draw(G, pos = pos, with_labels = True)
```

```python
# lets now fit our data in some models to see how well we can predict the classes 
# lets start with Logistic regression 
```

```python
# first we split our data into features and target 

#we remind our merged dataset
# print(df_class_and_features.head())

# we select classes 1 and 2 and we discard the unknown class 
data = df_class_and_features[(df_class_and_features['Class']=='1') | (df_class_and_features['Class']=='2')]

# features 
features =  data.drop(['ID','Class','Time step'], axis =1 )
#check 
# print(features)

# target 
target = data['Class']
#check 
print(target)

# we remind that  
# 1 means the transaction is illicit / illegal 
# 2 means the transaction is licit / legal 

# we rename the licit transactions from 0 to 2 
target = target.apply(lambda x: 0 if x == '2' else 1 )
#check 
# print(target)

# now we will have 
# 0 means the transaction is licit / legal 
# 1 means the transaction is illicit / illegal 

# so we split our data 
x_train, x_test, y_train, y_test = train_test_split(features,target,test_size=0.2,random_state=0)
```

```python
# time to fit our log regr model
reg = LogisticRegression(max_iter=500)
reg.fit(x_train,y_train)

# predictions 
pred = reg.predict(x_test)

# metrics for log regression 
prec,rec,f1,num = precision_recall_fscore_support(y_test,pred, average=None)
print("Logistic Regression")
print("Precision:%.3f \nRecall:%.3f \nF1 Score:%.3f"%(prec[1],rec[1],f1[1]))

# lets try macro-f1 also 
macro_f1 = f1_score(y_test,pred,average='macro')
print("Macro-Average F1 Score:",macro_f1)

# lets try micro-f1 also 
micro_f1 = f1_score(y_test,pred,average='micro')
print("Micro-Average F1 Score:",micro_f1)

# looks like our logistic regression does a descent job 
```

```python
# now lets try a random forest classifier 

rf = RandomForestClassifier(n_estimators=50, max_depth=100,random_state=0)
rf.fit(x_train,y_train)
pred = rf.predict(x_test)
prec,rec,f1,num = precision_recall_fscore_support(y_test,pred,average=None)
print("Random Forest Classifier")
print("Precision:%.3f \nRecall:%.3f \nF1 Score:%.3f"%(prec[1],rec[1],f1[1]))
macro_f1 = f1_score(y_test,pred,average='macro')
print("Macro-Average F1 Score:",macro_f1)
micro_f1 = f1_score(y_test,pred,average='micro')
print("Micro-Average F1 Score:",micro_f1)

#The random forest classifier improves further our scores 
```

```python
# we will finally try an MLP classifier 

MLP = MLPClassifier(random_state=0, max_iter=300, solver = 'adam', activation = 'relu')
MLP.fit(x_train,y_train)
pred = MLP.predict(x_test)
prec,rec,f1,num = precision_recall_fscore_support(y_test,pred,average=None)
print("MLP Classifier")
print("Precision:%.3f \nRecall:%.3f \nF1 Score:%.3f"%(prec[1],rec[1],f1[1]))
macro_f1 = f1_score(y_test,pred,average='macro')
print("Macro-Average F1 Score:",macro_f1)
micro_f1 = f1_score(y_test,pred,average='micro')
print("Micro-Average F1 Score:",micro_f1)
```

```python
# now we will try to explore what happens when we try to predict the unknown class 
# The unknown class has much more observations that the licit and illicit classes 
# lets use our best model in terms of scores the random forest classifier to see what is happening 

# we select the class 3 (unknown class) in this case from our initial dataframes 
Class3_ID = df_class_and_features.loc[(df_class_and_features['Class'] == 3), 'ID']
#CHECK 
# print(Class3_ID)
#check again how many unknown IDs i have 
# print(len(Class3_ID))

# picking the according values from all my dataframes 
df_edgelist_Class3_ID = df_edgelist.loc[df_edgelist['ID1'].isin(Class3_ID)]
df_classes_Class3_ID = df_classes.loc[df_classes['ID'].isin(Class3_ID)]
df_features_Class3_ID = df_features.loc[df_features['ID'].isin(Class3_ID)]

# Check 
# print(df_edgelist_Class3_ID.head())
# print(df_classes_Class3_ID.head())
# print(df_features_Class3_ID.head())
```

```python
# we create our predictions 
x_Class3_ID = df_features_Class3_ID.drop(columns=['ID', 'Time step'])
# we use our trained rf model 
pred_Class3_ID = rf.predict(x_Class3_ID.values)
print(np.unique(pred_Class3_ID))

# we replace the 'unknown' class with the predicted one 
df_classes_Class3_ID = df_classes_Class3_ID.copy()
df_classes_Class3_ID.loc[:, 'Class'] = pred_Class3_ID
# check 
print(df_classes_Class3_ID)
```

```python
# we merge class and features 
df_class_feature_Class3_ID = pd.merge(df_classes_Class3_ID, df_features_Class3_ID)
print(df_class_feature_Class3_ID.head())
```
